```python
import matplotlib.pyplot as plt
import seaborn as sns

# Residuos vs. valores ajustados
plt.figure(figsize=(10, 6))
sns.residplot(x=modelo.fittedvalues, y=modelo.resid, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Residuos vs. Valores Ajustados')
plt.show()
```

```python
import scipy.stats as stats

# Q-Q plot
plt.figure(figsize=(10, 6))
stats.probplot(modelo.resid, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar la base de datos
# df_cleaned = pd.read_csv('data.csv')  # AsegÃºrate de tener el archivo CSV en el directorio correcto

# AsegurÃ©monos de que no haya espacios adicionales en los nombres de las columnas
df_cleaned.columns = df_cleaned.columns.str.strip()

# Calcular la esperanza de vida promedio para cada paÃ­s
life_expectancy_avg = df_cleaned.groupby('Country')['Life expectancy'].mean().reset_index()

# Obtener los 5 paÃ­ses con la menor esperanza de vida promedio
top_5_menor = life_expectancy_avg.nsmallest(5, 'Life expectancy')

# Obtener los 5 paÃ­ses con la mayor esperanza de vida promedio
top_5_mayor = life_expectancy_avg.nlargest(5, 'Life expectancy')

print("PaÃ­ses con la menor esperanza de vida promedio:")
print(top_5_menor)

print("\nPaÃ­ses con la mayor esperanza de vida promedio:")
print(top_5_mayor)

# Graficar los resultados
plt.figure(figsize=(14, 8))

# GrÃ¡fica para los 5 paÃ­ses con la menor esperanza de vida promedio
plt.subplot(1, 2, 1)
sns.barplot(x='Life expectancy', y='Country', data=top_5_menor, palette='Reds_r')
plt.title('Top 5 paÃ­ses con menor esperanza de vida promedio')
plt.xlabel('Esperanza de vida promedio')
plt.ylabel('PaÃ­s')

# GrÃ¡fica para los 5 paÃ­ses con la mayor esperanza de vida promedio
plt.subplot(1, 2, 2)
sns.barplot(x='Life expectancy', y='Country', data=top_5_mayor, palette='Blues_d')
plt.title('Top 5 paÃ­ses con mayor esperanza de vida promedio')
plt.xlabel('Esperanza de vida promedio')
plt.ylabel('PaÃ­s')

plt.tight_layout()
plt.show()
```
